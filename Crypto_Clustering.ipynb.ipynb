```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
# Set the coinid column as index
# Copy the crypto names from the original data

df_scaled_data = pd.DataFrame(scaled_data, index=df_market_data.index, columns=df_market_data.columns)

# Display sample data
df_scaled_data.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_scaled_data)
    inertia.append(k_model.inertia_)
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_original_plot = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow_original
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4


---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_scaled_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
scaled_clusters = model.predict(df_scaled_data)

# Print the resulting array of cluster values.
print(scaled_clusters)
```

```python
# Create a copy of the DataFrame
df_scaled_data_predictions = df_scaled_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_scaled_data_predictions["ScaledCluster"] = scaled_clusters

# Display sample data
df_scaled_data_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
kmeans = KMeans(n_clusters=4)  
cluster_labels = kmeans.fit_predict(scaled_data)
df_market_data['cluster'] = cluster_labels
Cryptocurrency_scatter_plot = df_market_data.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by='cluster',
    hover_cols=['coin_id'],
    title="Cryptocurrency Clusters"
)

Cryptocurrency_scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)
# View the first five rows of the DataFrame. 
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])

df_pca.head(5)
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
total_explained_variance = explained_variance.sum()
print("Total Explained Variance:", total_explained_variance)
```

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.8950316570309841


```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
# Copy the crypto names from the original data
# Set the coinid column as index
df_pca = pd.DataFrame(pca_data, index=df_market_data.index, columns=['PC1', 'PC2', 'PC3'])
# Display sample data
df_pca.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in range(1, 12):
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(df_pca)  
    inertia_values.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve_dict = {'k': k_values, 'inertia': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_curve = pd.DataFrame(elbow_curve_dict)
```

```python
print("Length of k_values:", len(k_values))
print("Length of inertia_values:", len(inertia_values))
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_plot = df_elbow_curve.hvplot.line(x='k',y='inertia',title='Elbow Method For Optimal K',xticks='k_values')
pca_plot
```

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** Using fewer features to cluster data using K-Means can have a significant impact. Reducing the dimensionality of the dataset might lead to faster computation and simpler models. It can also help to avoid the "curse of dimensionality" which can make clustering less effective as the distance between points becomes less meaningful in high-dimensional spaces. However, using fewer features may also result in the loss of important information that could lead to less accurate or meaningful clusters. The key is to balance the number of features to retain enough information while avoiding unnecessary complexity.
